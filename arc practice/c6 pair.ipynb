```python
import arc
from arc import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
ns = np.arange(60,80,2)
atomDistance = 7

Rb_c6s = []
Cs_c6s = []

V_Rb = []
V_Cs = []

for n in ns:
    calculation = arc.PairStateInteractions(Rubidium(), n, 0, 0.5, n, 0, 0.5, 0.5, 0.5, interactionsUpTo= 2)

    c6 = calculation.getC6perturbatively(0,0,10,3e10)

    V_Rb.append((-c6 / atomDistance**6) * 1.0e3)

    Rb_c6s.append(-c6)
    calculation = arc.PairStateInteractions(Caesium(), n, 0, 0.5, n, 0, 0.5, 0.5, 0.5, interactionsUpTo= 2)

    c6 = calculation.getC6perturbatively(0,0,10,3e10)
    Cs_c6s.append(-c6)
    V_Cs.append((-c6 / atomDistance**6) * 1.0e3)
```

```python
plt.figure()
plt.plot(ns,Rb_c6s, label = r'$|l=0, j=0.5, m=0.5 \rangle_{Rb} $', marker='.')
plt.plot(ns,Cs_c6s, label = r'$|l=0, j=0.5, m=0.5 \rangle_{Cs} $', marker='.')
plt.legend()
plt.ylabel('$C_6$ $(GHz \mu m^6)$')
plt.xlabel('n')
plt.show()
```

```python
plt.figure()
plt.plot(ns,V_Cs, label = r'$|l=0, j=0.5, m=0.5 \rangle_{Cs} $', marker = '.')
plt.plot(ns,V_Rb, label = r'$|l=0, j=0.5, m=0.5 \rangle_{Rb} $', marker = '.')
plt.legend()
plt.ylabel('$V$ $(MHz)$')
plt.xlabel('n')
plt.show()
```
