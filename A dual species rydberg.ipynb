```python
%matplotlib inline
import matplotlib.pyplot as plt  # Import library for direct plotting functions
import numpy as np  # Import Numerical Python
from IPython.display import display, HTML  # Import HTML for formatting output


from arc import *
```

```python
atom1 = Caesium()
atom2 = Rubidium()
ns = np.arange(60,80)

l0 = 0
j0 = 0.5
mj0 = 0.5
# Target State
theta = 0
# Polar Angle [0-pi]
phi = 0
# Azimuthal Angle [0-2pi]
dn = 0
# Range of n to consider (n0-dn:n0+dn)
deltaMax = 25e9  # Max pair-state energy difference [Hz]

Rb_C6s = []
Cs_C6s = []
for n in ns:
    calculation = PairStateInteractions(
    Rubidium(), n, l0, j0, n, l0, j0, mj0, mj0
    )
    C6 = calculation.getC3perturbatively(theta, phi, dn, deltaMax)
    
    Rb_C6s.append(C6)

    calculation = PairStateInteractions(
    Caesium(), n, l0, j0, n, l0, j0, mj0, mj0
    )
    C6 = calculation.getC3perturbatively(theta, phi, dn, deltaMax)

    Cs_C6s.append(C6)


plt.figure()
plt.plot(ns, Rb_C6s, label ='Rb')
plt.plot(ns, Cs_C6s, label ='Cs')
plt.legend()
plt.show()
```
